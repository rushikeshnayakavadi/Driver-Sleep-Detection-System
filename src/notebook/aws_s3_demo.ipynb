```python
import boto3
import os

# Initialize S3 client
s3 = boto3.client("s3")
bucket_name = "driver-sleep-detection-data"

def upload_folder_to_s3(local_folder, s3_folder):
    if not os.path.exists(local_folder):
        print(f"Local folder '{local_folder}' does not exist!")
        return
    
    for root, _, files in os.walk(local_folder):
        for file in files:
            local_path = os.path.join(root, file)
            
            # Preserve subfolder names (Closed, no_yawn, Open, yawn)
            relative_path = os.path.relpath(local_path, local_folder)  
            s3_path = os.path.join(s3_folder, relative_path).replace("\\", "/")  # Convert to S3 format
            
            print(f"ðŸ“¤ Uploading {local_path} â†’ s3://{bucket_name}/{s3_path}")
            s3.upload_file(local_path, bucket_name, s3_path)
    
    print(f"Upload completed for '{local_folder}'")

# Upload Train and Test folders with exact subfolder names
upload_folder_to_s3("dataset_new/train", "train")
upload_folder_to_s3("dataset_new/test", "test")
```
